Notes on Pandas

Attributes vs. Methods

Attributes:
    describes the features of the object (explains what is already there
    .values
    .index
    .dtype
    .shape

Methods:
    Method is a function passed to the object that can have its own paramaters aka ()
    Always end in a ()
    Can chain methods together via method chaining
    Methods always do something to the value

```python
#Examples of Attributes and Methods

#Attributes
df['score_phrase'].values
df['score_phrase'].index
print(df['score'].dtype) #Seems extremely useful
print(df.shape) #another good one for understanding the size


#Methods
print(df.describe())
print(df['score'].sum())
print(df['score'].mean())
```

```python
#Series Attribtues
print(df['title'].head().str.len())

#two things going on
    # .str allows you to read the series as a strings
    # then take the length of them

#Selects rows 0 - 4, was interested in what the 6 character title game was
df[0:5]
```

```python
df['url'].is_unique
```

```python
#sort by numbers
df[['score','title']].sort_values(ascending = False, by = 'score')

#sort by abc
df[['score','title']].sort_values(ascending = True, by = 'title')

#using this sort by keep the series intact, however if you set inplace = True, will permenantly modify them
#Can also sort by index via sort_index()
```

```python
# creates a boolean list that can be used to filter data
#df['platform'].isin(['PlayStation 3'])

#yields only PlayStation 3 Games
df[df['platform'].isin(['PlayStation 3'])]
```

```python
#Can do multiple filters as well using isin

#great & editors choice & Sports

df[df['score_phrase'].isin(['Great']) & 
   df['editors_choice'].isin(['Y']) & 
   df['genre'].isin(['RPG']) ]

#is case sensative
```

```python
#Can also use ~ to do "not in"

df[~df['score_phrase'].isin(['Great']) & 
   ~df['editors_choice'].isin(['Y']) & 
   ~df['genre'].isin(['RPG']) ]
```

```python
#indexing

df[100:105] #returns 5 rows, remembering that 
```

```python
#an alternative to just passing it
print(df.get(['title','url']).head())

print(df[['title','url']].head())
```
