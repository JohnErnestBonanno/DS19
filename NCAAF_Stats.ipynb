```python
from sportsreference.ncaaf.boxscore import Boxscore
from sportsreference.ncaaf.schedule import Schedule
from sportsreference.ncaaf.conferences import Conferences

import pandas as pd
```

```python
#Because teams isn't working, need to get team names from conferences
```

```python
# creates conferences object
conferences = Conferences()

#makes a dictionary of all the teams & their conferences
conf_dict = conferences.team_conference
```

```python
#converts dictionary into a data frame
conf_df = pd.DataFrame(conf_dict.items(), columns = ['team','conference'])

#makes a list of all NCAAF teams for iteration
teams = list(conf_df['team'])
```

```python
#subset ACC teams for testing puproses
acc_teams = list(conf_df['team'][conf_df['conference']=='acc'])
```

```python
for team in acc_teams:
    print(team)
```

```python
df_master = pd.DataFrame()

for team in acc_teams:
    schedule = Schedule(team,2018)
    for game in schedule:
        print(game.boxscore_index)
        df_individual = game.boxscore.dataframe
        df_master = df_master.append(df_individual)
```

```python
#code to actually use

#teams

year_list = []
for i in range(2000,2019):
    year_list.append(i)
```

```python
#data to actually use, but code will take a while to run
df_master = pd.DataFrame()

for team, year in zip(teams, year_list):
    schedule = Schedule(team,year)
    for game in schedule:
        print(game.boxscore_index)
        df_individual = game.boxscore.dataframe
        df_master = df_master.append(df_individual)
```
